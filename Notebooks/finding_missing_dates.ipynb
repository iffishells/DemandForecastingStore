# Loading libraries

```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
def find_missing_date(df):
    df['date'] = pd.to_datetime(df['date'],format='%Y-%m-%d')
    max_date = df['date'].max()
    min_date = df['date'].min()
    date_range = pd.date_range(min_date, max_date)
    missing_dates = date_range[~date_range.isin(df['date'])]
    return missing_dates ,min_date,max_date


```

```python
path_list = glob.glob('../processedData/Sku_shop_based_data/*.csv')
missing_dates_information = {
    'fileName': [],
    'missing_dates':[],
    'Number_of_missing_dates':[],
    'data_start_date':[],
    'data_end_date':[]
}
for path in path_list:
    df = pd.read_csv(path)
    missing_dates,start_date , end_date = find_missing_date(df)
    lenght_of_missing_dates = len(missing_dates)
    missing_dates_information['fileName'].append(path.split('/')[-1].split('.')[0])
    missing_dates_information['missing_dates'].append(missing_dates)
    missing_dates_information['Number_of_missing_dates'].append(lenght_of_missing_dates)
    missing_dates_information['data_start_date'].append(start_date)
    missing_dates_information['data_end_date'].append(end_date)
    # break
    
missing_dates_df = pd.DataFrame.from_dict(missing_dates_information)
missing_dates_df.to_csv('../processedData/meta_information/Missing_dates.csv',index=False)
```
